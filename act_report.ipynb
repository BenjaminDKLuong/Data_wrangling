```python
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
df = pd.read_csv('twitter_archive_master.csv')
```

# Insights and Visualization

```python
# plot Number of pictures in a tweet
ind = df.img_num.value_counts().index
df.img_num.value_counts()[ind].plot(kind = 'bar',figsize = (19,9))
plt.xlabel('Number of pictures')
plt.ylabel('Occurences')
plt.xticks([0,1,2,3] ,['1','2','3','4'])
plt.title('Number of pictures in a tweet');
```

## Insight 1:

**By looking at the graph above, I see that most tweets have one picture.**

```python
# scatterplot to see relationship between retweet_count and favorite_count
df.plot(x = 'favorite_count' , y = 'retweet_count', kind = 'scatter', figsize = (19,19))
plt.xlabel('Favorite count')
plt.ylabel('Retweet count')
plt.title('Retweet count and Favorite count');
```

## Insight 2:

**By looking at the graph above, I see that favorite_count has positive correlation with retweet_count.  It means that the more people favorite a tweet, the more people will retweet it. Moreover, there are people who like to retweet instead of favoriting a tweet.  That's explain the vertical line at favorite count equal zero.**

```python
# plot Number of pictures in a tweet
ind = df.dog_stage.value_counts().index
df.dog_stage.value_counts()[ind].plot(kind = 'bar',figsize = (19,9))
plt.ylabel('Occurences')
plt.title('Stage of dogs');
```

## Insight 3:

**By looking at the graph above, I see that people post their dogs at pupper stage the most.**
