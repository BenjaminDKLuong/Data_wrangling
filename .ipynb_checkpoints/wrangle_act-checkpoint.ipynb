### Download  data from Twitter API

```python
# create a list of tweet_id for downloading
tweet_id_list = df_archive.tweet_id
tweet_id_list.head()
```

```python
# get access to twitter api
import tweepy

consumer_key = ''
consumer_secret = ''
access_token = ''
access_secret = ''

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth, wait_on_rate_limit = True, wait_on_rate_limit_notify = True)
```

```python
#download tweets
df_tweet_api = []
errors = []
count = 0
for tweet_id in tweet_id_list:        
    try:
        #keeping tracking where we are 
        count += 1
        print (str(count)+"_"+str(tweet_id))
        
        tweet = api.get_status(tweet_id, tweet_mode = 'extended')
        retweet_count = tweet.retweet_count
        favorite_count = tweet.favorite_count
        
        #append the tweet into df_tweet_api
        df_tweet_api.append({'tweet_id': int(tweet_id),
                             'retweet_count': int(retweet_count),
                             'favorite_count': int(favorite_count)})

    except:
        print (str(tweet_id) + "_is_deleted")
        errors.append(tweet_id)
```

```python
df_tweet_api[:2]
```

```python
# how many tweet we downloaded
len(df_tweet_api)
```

```python
# how many errors we got (tweets are deleted)
len(errors)
```

```python
#save df_tweet_api with json into tweet_json.txt
import json

with open('tweet_json.txt', 'w') as file:
    json.dump(df_tweet_api, file)
```

```python
# open json file: tweet_json.txt
with open('tweet_json.txt') as json_file:
    df_api = pd.read_json(json_file)
    
df_api.head()
```

### Quality: issues with content. Low quality data is also known as dirty data.


#### Define and Code

```python
# drop low quality data columns: in_reply_to_status_id, in_reply_to_user_id, retweeted_status_id, retweeted_status_user_id, retweeted_status_timestamp
df_archive_clean = df_archive_clean.drop(['in_reply_to_status_id', 'in_reply_to_user_id', 'retweeted_status_id', 'retweeted_status_user_id', 'retweeted_status_timestamp'], axis =1)
```

#### Define and Code

```python
#timestamp: datatype: datetime
df_archive_clean.timestamp = pd.to_datetime(df_archive_clean.timestamp)
```

#### Define and Code

```python
# tweet_id: change datatype to string
df_archive_clean.tweet_id = df_archive_clean.tweet_id.astype(str)
df_predictions_clean.tweet_id = df_predictions_clean.tweet_id.astype(str)
df_api_clean.tweet_id = df_api_clean.tweet_id.astype(str)
```

#### Define and Code

```python
#img_num : datetype: category (df_predictions_clean)
df_predictions_clean.img_num = df_predictions_clean.img_num.astype('category')
```

#### Define and Code

```python
#replace None to NAN for name, doggo, floofer, pupper, and puppo columns
df_archive_clean.name = df_archive_clean.name.replace('None', np.nan)
df_archive_clean.doggo = df_archive_clean.doggo.replace('None', np.nan)
df_archive_clean.floofer = df_archive_clean.floofer.replace('None', np.nan)
df_archive_clean.pupper = df_archive_clean.pupper.replace('None', np.nan)
df_archive_clean.puppo = df_archive_clean.puppo.replace('None', np.nan)
```

#### Define and Code

```python
# expanded_urls : fill NAN
mask = df_archive_clean.expanded_urls.isnull()
df_archive_clean.loc[mask,'expanded_urls'] = "https://twitter.com/dog_rates/status/"+ df_archive_clean.tweet_id.astype(str)
```

#### Define and Code

```python
# fix rating_numerator: by extract it from text column
# extract value from text column to a dummy column
df_archive_clean['new_num'] = df_archive_clean.text.str.extract('((\d+(\.\d+)?)/(\d+))',expand = True)
# replace rating_numerator and rating_denominator columns with new values
df_archive_clean['rating_numerator'],df_archive_clean['rating_denominator'] = df_archive_clean['new_num'].str.split('/',1).str
# drop the new_num column
df_archive_clean = df_archive_clean.drop('new_num',axis =1)
```
